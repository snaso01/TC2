# Simulacin en Phython

```python

# Libreras externas NumPy, SciPy y Matplotlib
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, freqs


# Librera de TC2, esta la vas a usar mucho
from pytc2.sistemas_lineales import pzmap, bodePlot

w0 = 2*np.pi*10e3
qq = 20
k = 10

num =  [k*w0/qq, 0];
den =  [1, w0/qq, w0**2];

w = np.logspace(3, 6, 500)  # frecuencia real de 1kHz a 1MHz (rad/s)
f_hz = w / (2 * np.pi)

_, h = freqs(num, den, w)
phase = np.angle(h, deg=True)

plt.figure()
plt.semilogx(w_hz, np.abs(h))
plt.grid(True, which='both')
plt.title(f'Respuesta en magnitud (K = {k}, f0 = 10KHz y Q = 20)')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('|H(j)|')
plt.show()

plt.subplot(2, 1, 2)
plt.semilogx(w_hz, phase, 'g')
plt.title(f'Fase de respuesta en frecuencia')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Fase []')
plt.grid(True, which='both')
```
