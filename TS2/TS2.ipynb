```python
# Librer√≠as externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np


# Librer√≠a de TC2, esta la vas a usar mucho
from pytc2.sistemas_lineales import pzmap, bodePlot

w0 = 1
q_param = [0.5, 1, 10, 100]
k = 10

for q in range(len(q_param)):

    num =  [k*w0**2];
    den =  [1, w0/(q_param[q]), w0**2];

    my_tf = TransferFunction(num, den)

    bodePlot(my_tf, fig_id=1, filter_description = 'Q={:.1f}'.format(q_param[q]))

    pzmap(my_tf, fig_id=2, filter_description = 'Q={:.1f}'.format(q_param[q])) #S plane pole/zero plot
```

#### Mediante estas simulaciones podemos confirmar que tomando Vo como salida, el circuito se comporta como un filtro pasabajos de 2do orden y que cuanto m√°s grande sea el factor de calidad (Q), m√°s selectivo se vuelve el filtro.

#### Para poder obtener un circuito pasabanda se toma como transferencia Vo1/Vo, es decir se toma como salida Vo1
